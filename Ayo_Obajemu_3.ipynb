Task 3: Comparative Visualization (Matplotlib)

```python
import matplotlib.pyplot as plt
```

```python
for platform in df['platform'].unique():
    plt.hist(df[df['platform'] == platform]['views'], alpha=0.5, label=platform)

plt.title("Distribution of Views by Platform")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.scatter(df['views'], df['likes'])
plt.title("Relationship Between Views and Likes")
plt.xlabel("Views")
plt.ylabel("Likes")
plt.show()
```

Task 4: Trend & Distribution Analysis

```python
plt.plot(df.index, df['views'])
plt.title("Trend of Views Over Time")
plt.xlabel("Index")
plt.ylabel("Views")
plt.show()
```

```python
plt.hist(df['views'])
plt.title("Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.show()
```
