```python
likes_column = df["likes"]
```

```python
import numpy as np

likes_array = likes_column.to_numpy()
```

```python
mean_value = np.mean(likes_array)
median_value = np.median(likes_array)
min_value = np.min(likes_array)
max_value = np.max(likes_array)
std_value = np.std(likes_array)
```

```python
print("Mean:", mean_value)
print("Median:", median_value)
print("Minimum:", min_value)
print("Maximum:", max_value)
print("Standard Deviation:", std_value)
```

```python
plt.figure()
for platform in df["platform"].unique():
    plt.hist(
        df[df["platform"] == platform]["likes"],
        alpha=0.5,
        label=platform
    )

plt.title("Distribution of Likes by Platform")
plt.xlabel("Number of Likes")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(df["likes"], df["shares"])
plt.title("Relationship Between Likes and Shares")
plt.xlabel("Likes")
plt.ylabel("Shares")
plt.show()
```

```python
plt.figure()
plt.plot(df["likes"])
plt.title("Trend of Likes Over Time")
plt.xlabel("Post Index")
plt.ylabel("Likes")
plt.show()
```
