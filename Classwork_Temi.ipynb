```python
import numpy as np
numerical_column = df["likes"]
numerical_column
```

```python
num_array = numerical_column.to_numpy()
num_array
```

```python
mean_value = np.mean(num_array)
median_value = np.median(num_array)
min_value = np.min(num_array)
max_value = np.max(num_array)
std_value = np.std(num_array)

mean_value, median_value, min_value, max_value, std_value
```

```python
#Task 4
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(num_array)
plt.title("Line Chart of Numerical Column")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
plt.figure()
plt.bar(range(len(num_array)), num_array)
plt.title("Bar Chart of Numerical Column")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

#Insights

1. The selected numerical column shows noticeable variation across the dataset, indicating changes over time or index.
2. The line chart reveals trends and fluctuations, suggesting possible peaks and drops in the values.
3. The bar chart highlights differences between individual observations, making it easier to compare values visually.
